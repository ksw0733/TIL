#### 메론 순위 전체 데이터 가져오기
- 출력은 DataFrame

```python
driver = webdriver.Chrome("./chromedriver.exe")
time.sleep(1)
url="https://www.melon.com/chart/day/index.htm"
driver.get(url)
driver.implicitly_wait(10)

html = driver.page_source
soup= bs(html, "html.parser")
driver.quit()
tbody_soup = soup.find('tbody')
tr_soup = tbody_soup.find_all("tr")

tr_list=[]
for tr in tr_soup:
    rank=tr.find("span", class_="rank").get_text().replace("\n", "")
    title=tr.find("div", class_="ellipsis rank01").get_text().replace("\n", "")
    art=tr.find("span", class_="checkEllipsis").get_text().replace("\n", "")
    ell=tr.find("div", class_="ellipsis rank03").get_text().replace("\n", "")
    tr_list.append([rank, title, art, ell])
    
df=pd.DataFrame(tr_list, columns=["순위","곡제목", "가수명", "앨범"])
df
```
