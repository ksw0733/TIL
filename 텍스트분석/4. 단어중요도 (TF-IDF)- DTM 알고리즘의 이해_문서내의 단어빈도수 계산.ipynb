### Part 1-5 (DTM: 단어별 count)

```python
######################################
## 1개의 컬럼만 출현횟수를 먼저 계산하여 봅니다.
## txtList는 아래와 같습니다.
##  [['파이썬', '차트', '파이썬', '머신러닝'], 
##    ['차트', '파이썬', 'R', '차트'], 
##   ['R', '분석', '시각화']]
#####################################
## 참고, 아래의 경우 uniqueList의 첫단어를 1번행에서만 확인한 결과임. [1행 1열의 값을  보여줌]

col=uniqueList[0]   # uniqueList의 첫번째 글자인 R 입력
print('▶ 검색단어==> %s'%(col))

row=txtList[0]   # 1번행에서만 확인하기
print(f'▶ 검색대상 1행 ==> {row}')

cnt = 0
for item in row:
    if col == item:
        cnt += 1

print('▶▶▶ 검색단어: %s   검색된횟수: %d' %(col,cnt))
```

```python
### -- DTM 계산
######################################
### 컬럼자료의 출현횟수를 계산함.
### [방법1] 직접작성
### 위의 코드를 다중 for로 완성
###################################

result=[]   # 각 문장별/단어 빈도수 저장

for col in uniqueList:
    list_=[]
    for row in txtList:
        cnt = 0
        for item in row:
            if col == item:
                cnt += 1

        list_.append(cnt)

    result.append(list_)

print(result)
print("========================================")
df = pd.DataFrame(result)
display(df)

df=df.T  # DataFrame에서 행/열 위치 교환
display(df)

df.columns=uniqueList
display(df)
```

```python
######################################
### 참고: 장바구니분석(연관분석)에 사용하는 자료는 자료의 출현여부만 확인함.
### [방법1] 직접작성
#####################################

result=[]
for data in txtList:
    
    list_=['True' if x in data  else 'False' for x in uniqueList]
    result.append(list_)
print(np.array(result))

pd.DataFrame(result,columns=uniqueList)
```

```python
######################################
### 참고: 장바구니분석(연관분석)에 사용하는 자료는 자료의 출현여부만 확인함.
### [방법2] transactionEncoder 모듈 사용
#####################################

from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te.fit(txtList)
print(te.columns_)

print('--------------------------------------------')

te_ary = te.fit(txtList).transform(txtList)
print(te_ary)

df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

### [Part2] DTM 제작간단히code
- https://wikidocs.net/31698
- 위의 위키문서는 텍스트분석에 대하여 아주 잘 서술된 자료입니다. 
- 이 자료를 기반으로 2번에서 TF-IDF를 제작하여 보겠습니다.

```python
# 데이터에서 필드(고유단어) 추출하기
txt=['파이썬 차트 파이썬 머신러닝',
      '차트 파이썬 R 차트',
       'R 분석 시각화'
      ]

vocab = list(set(w for doc in txt for w in doc.split()))
print(vocab)

vocab.sort()
print(vocab)
```

```python
result = []
N=len(txt)
for i in range(N): # 각 문서에 대해서 아래 명령을 수행
    result.append([])
    d = txt[i]
    for j in range(len(vocab)):
        t = vocab[j]        
        result[-1].append(d.count(t))

tf_ = pd.DataFrame(result, columns = vocab)
tf_
```
